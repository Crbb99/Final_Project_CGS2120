```python
import math

# Function Definitions

def add(a, b): # Returns the sum of a and b
    return a + b

def subtract(a, b): # Returns the difference of a and b
    return a - b

def multiply(a, b): # Returns the product of a and b
    return a * b

def divide(a, b): # Returns the quotient of a divided by b and handles division by zero
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Division by zero is undefined."

def power(a, b): # Returns a raised to the power of b
    return a ** b

def square_root(a): # Returns the square root of a and handles negative input errors
    if a < 0:
        return "Error: Cannot compute square root of a negative number."
    return math.sqrt(a)

# User Interaction Function

def calculator(): # Interactive calculator function for user input and operation selection
    while True:
        print("\nSelect an operation:")
        print("1 for Addition (+)")
        print("2 for Subtraction (-)")
        print("3 for Multiplication (*)")
        print("4 for Division (/)")
        print("5 for Exponentiation (^)")
        print("6 for Square Root (√)")
        print("7 to Exit")

        choice = input("Enter the number corresponding to your choice: ")

        if choice == '7':
            print("Exiting the calculator. Goodbye!")
            break

        if choice in ['1', '2', '3', '4', '5']:
            try:
                num1 = float(input("Enter the first number: "))
                num2 = float(input("Enter the second number: "))
            except ValueError:
                print("Invalid input! Please enter numeric values.")
                continue

        if choice == '6':
            try:
                num1 = float(input("Enter the number: "))
            except ValueError:
                print("Invalid input! Please enter a numeric value.")
                continue

        # Map user choice to the corresponding function
        if choice == '1':
            print(f"Result: {add(num1, num2)}")
        elif choice == '2':
            print(f"Result: {subtract(num1, num2)}")
        elif choice == '3':
            print(f"Result: {multiply(num1, num2)}")
        elif choice == '4':
            print(f"Result: {divide(num1, num2)}")
        elif choice == '5':
            print(f"Result: {power(num1, num2)}")
        elif choice == '6':
            print(f"Result: {square_root(num1)}")
        else:
            print("Invalid input! Please select a valid operation.")

# Testing the Functions

def test_functions(): # Test cases for all arithmetic functions
    print("Testing addition:", add(2, 3) == 5)
    print("Testing subtraction:", subtract(7, 3) == 4)
    print("Testing multiplication:", multiply(4, 3) == 12)
    print("Testing division:", divide(21, 3) == 7)
    print("Testing division by zero:", divide(1, 0) == "Error: Division by zero is undefined.")
    print("Testing exponentiation:", power(4, 2) == 16)
    print("Testing square root:", square_root(144) == 12)
    print("Testing square root of negative:", square_root(-25) == "Error: Cannot compute square root of a negative number.")

# Main Execution
if __name__ == "__main__":
    print("Welcome to Caelan Beebe's Command-Line Calculator!")
    test_functions()  # Run tests
    calculator()  # Start the calculator
```

Functions are essential for efficiently organizing and simplifying code. They encourage modularity and lessen repetition by encapsulating particular operations into reusable blocks. For instance, the program is more structured and debug-friendly because the arithmetic functions each handle a single operation. Any logic that has to be adjusted can be done so in one place without impacting the program as a whole. By using meaningful names and docstrings to explicitly clarify each function's purpose, this method also improves readability. The use of a centralized calculator() function for user interaction helps to streamline the process by delegating operations to the appropriate arithmetic functions. This separation of responsibilities increases maintainability because changes to the user interface logic do not affect the computing logic.
